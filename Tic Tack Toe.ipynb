```python
def intro():
    global x
    global o
    class Player:
        def __init__(self,username,var,int):
            self.username = username
            self.var = var
            self.int = int
    x = Player(input("Whos going first? "),"x",1)
    o = Player(input("Great now who is the challenger? "),"o",2)
tic= {
    "A1":0,
    "A2":0,
    "A3":0,
    "B1":0,
    "B2":0,
    "B3":0,
    "C1":0,
    "C2":0,
    "C3":0,
}
def showb():
    string_to_print = ''
    for index, value in enumerate(tic.values()):
      if value == 1:
        string_to_print += 'x'
      elif value == 2:
        string_to_print += 'o'
      else:
        string_to_print += ' '
      if (index + 1) % 3 == 0:
        string_to_print += '\n'
      else:
        string_to_print += ' | '
    print(string_to_print)
def hturn(a):
    global tack
    tack=0
    while True:
        showb()
        tack = input("okay, " + a.username + " you're up! ")
        if tack.capitalize() in tic:
            tic[tack.capitalize()]= a.int
            wincheck()
            break
        else:
            print("please try again A-C and 1-3")
            continue
def wincheck():
    global W 
    W = False
    if tic["A1"] == tic["A2"] and tic["A1"] == tic["A3"]:
        if tic["A1"]== False:
            return ("next turn8")
        else:
            print("Top Row winner")
            W = True
    elif tic["A1"] == tic["B2"] and tic["A1"] == tic["C3"]: 
        if tic["A1"]== False:
            return ("next turn2")
        else:
            W = True
            print("Left Cross Winner")
    elif tic["A3"] == tic["B2"] and tic["A3"] == tic["C1"]: 
        if tic["A3"]== False:
            return ("next turn1")
        else:
            W = True
            print("Right Cross Winner")    
    elif tic["B1"] == tic["B2"] and tic["B1"] == tic["B3"]:
        if tic["B1"]== False:
            return ("next turn7")
        else:
            W = True
            print("Mid Row Winner")
    elif tic["C1"] == tic["C2"] and tic["C1"] == tic["C3"]:
        if tic["C1"]== False:
            return ("next turn6")
        else:
            W = True
            True("Bottom Row Winner")
    elif tic["A1"] == tic["B1"] and tic["A1"] == tic["C1"]: 
        if tic["A1"]== False:
            return ("next turn5")
        else:
            W = True
            print("Left Column Winner") 
    elif tic["A2"] == tic["B2"] and tic["A2"] == tic["C2"]:
        if tic["A2"]== False:
            return ("next turn4")
        else:
            W = True
            print("Mid Column Winner")
    elif tic["A3"] == tic["B3"] and tic["A3"] == tic["C3"]:
        if tic["A3"]== False:
            return ("next turn3")
        else:
            W = True
            print("Right Column Winner")
    else:
        return ("Draw")
def main_game(x,o):
    for i in range(9):
        hturn(x)
        if W == True:
            print("Congrat ",x.username," you won in ",i+1," turns!\nplay again?" )
            break
        else:
            hturn(o)
            if W == True:
                print("Congrat ",o.username," you won in ",i+1," turns!\nplay again?" )
                break
            else:
                continue
#to run the game
intro()
main_game(x,o)
```
